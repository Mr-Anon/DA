```python
import time
start_time = time.time()

def get_1_candidates(transactions):
    
    ans = []
    for i in transactions:
        ans = ans + i
    ans = list(set(ans))
    ans = [[i] for i in ans]
    return ans
    

def getData(filename):
    
    with open(filename, 'r') as file:
        data = file.read().replace('\n', '')
    file.close()
    
    data = data.split('-2')
    
    for i in range(len(data)):
        data[i] = data[i].split('-1')
        data[i] = data[i][:-1]
        data[i] = list(map(int, data[i]))
        
    sm=0
    for i in data: sm = sm + len(i)
    print(f'The dataset has {len(data)} transactions with average {sm/len(data)} items per transaction.')
    return data

dataset = getData('SIGN.txt')
get_1_candidates(dataset)
# print(f"TIME: {(time.time() - start_time)} seconds")
```
